We start with a dataset from a company called checkars (now kavak). Let's explore the dataset together using Pandas.

The dataset contains the information of 10.000 second-hand cars. There are 9 different columns:

- Make (Car brand, example: Ford)
- Model (The Model of the Car, example: Focus)
- Year (The Year in which the car was build, example: 2012)
- Variant (The car model version showing the PS, example: 1.6 Trendline)
- Kms (The kilometers the cars has been driven, example: 90000)
- Price (The offered price for the car, example: 10000)
- Doors (How many doors the car has, example: 4)
- Kind (Type of car, example: Pick-Up)
- Location (Where the car is located, example: Buenos Aires)

## ***.iloc[]***

**The .iloc[] method**

**iloc** stays for **"integer location"**

We can use iloc to select and retrieve data from a DataFrame based on integer-based indexing. For example, to access the third row, we can do:

````
#Notice that the first row is the 0-indexed
df_cars.iloc[2]
````

````
# method 1: access row 4
df_cars.iloc[3]
````

````
# selecting continuous rows using integer location: accessing row 1 to 5.
df_cars.iloc[1:5]
````

````
# selecting discrete rows using integer location --> Pass list of indexes.
# access row 1, 2, 4
df_cars.iloc[[1,2,4]]
````

````
#iloc accepts the second "argument" - column number!
df_cars.iloc[1,2]
````

````
# selecting certain rows and columns using iloc.
# example: row --> 1, 2, 4 of column 1, 7
df_cars.iloc[[1,2,4],[1,7]]
df_cars.iloc[[1,2],[1,7]]
````

Do you really need to remember or count column number every time?

Not really! To locate data using column names we can use **loc** method:

## ***.loc[]***

````
# selecting certain row & cols using loc.
# Select row 1, 2, 4 of column 'Year', 'Make'
df_cars.loc[[1,2,4],['Year','Make']]
````

````
df_cars.loc[1,["Kind"]]
````

Small pandas summary:
- head - top N rows
